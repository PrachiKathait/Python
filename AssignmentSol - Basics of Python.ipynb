# QUES 6

```python
def IsPrime(n):
    for i in range(2, n//2 + 1):
        if n%i==0:
            return 0
    return 1
def IsOddOrEven(n):
    if n%2==0:
        return 0 #EVEN
    return 1 #ODD
import random as r
a=[]
for i in range(100):
    a.append(r.randint(100,900))
print(a)
Prime=[]
Odd=[]
Even=[]
for i in range(0,100):
    if IsPrime(a[i])==1:
        Prime.append(a[i])
    if IsOddOrEven(a[i])==0:
        Even.append(a[i])
    if IsOddOrEven(a[i])==1:
        Odd.append(a[i])
print(f"Prime List : {Prime}")
print(f"Total numbers : {len(Prime)}")
print(f"Even List : {Even}")
print(f"Total numbers : {len(Even)}")
print(f"Odd List : {Odd}")
print(f"Total numbers : {len(Odd)}")
```

```python
import random as r
def sMarks():
    S=[]
    total=[]
    n=r.randint(2,6)
    for i in range(n):
        D={}
        for i in range(5):
            D[i]=r.randint(40,100)
        S.append(D)
        total.append(sum(list(D.values())))
    return S,total
s, t = sMarks()
print(f"{s}\n")
print(t)
print(max(t))
print(min(t))
print(sum(t)/len(s))
```
